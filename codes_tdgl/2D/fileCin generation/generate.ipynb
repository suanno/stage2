```python
import os
import shutil
import time
import numpy as np
from matplotlib import pyplot as plt
```

```python
dt = 0.1
tspan = 2000
N = int(tspan/dt)
Cs = np.zeros(N)
ts = np.linspace(0,tspan,N)
```

```python
# Create evolution directory
simul_name = "t=100dt=1e-1C=1"

file_simul = ".saves/" + simul_name + "/" 
name = "fileCin.dat"
dir = "../" + file_simul + name
# Copy files
f = open(dir, "w")
for i in np.arange(0,N):
    T = 100*dt
    Cs[i] = 0 + 1*np.sin(2*np.pi*ts[i]/T)
    Cs[i] = 1
    f.write(str(ts[i]) + " " + str(Cs[i]) + "\n")
f.close()

plt.plot(ts, Cs)
plt.title("dt = " + str(dt))
plt.ylabel("C(t)")
plt.xlabel("t")
```

```python
import os
import shutil
import time
import numpy as np
from matplotlib import pyplot as plt
```

```python
dt = 0.01
tspan = 10
N = int(tspan/dt)
Cs = np.zeros(N)
ts = np.linspace(0,tspan,N)
```

```python
# Create evolution directory
name = "fileCin.dat"
dir = "../" + name
# Copy files
f = open(dir, "w")
for i in np.arange(0,N):
    Cs[i] = np.sin(2*np.pi*ts[i])
    f.write(str(ts[i]) + " " + str(Cs[i]) + "\n")
f.close()

plt.plot(ts, Cs)
```

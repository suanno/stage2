```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
import shutil
import time
```

```python
N = 10000000
dir_save = ".saves/" + "Crank-Nicolson fixed T_over_dt=4 Cave=0 u0=1/" 
dt = 0.008
tmax = 5000
imax = int(tmax/dt)
```

```python
# Create plot object
fig, ax = plt.subplots()
ax.set_xscale("log")
ax.set_xlabel("t")
ax.set_ylabel("u(nT)")
ax.set_yscale("log")
ax.grid()
ax.set_title("Crank-Nicolson vary dt Cave=0 u0=1 T=2")
```

```python
# Get all the simulations in the selected folder
for folder in sorted(os.listdir(dir_save), reverse=True):
    dir = dir_save + folder
    print(dir)
    # Get initial value
    f = open(dir + "/fileAveout.dat", "r")
    label = f.readline().strip('\n')
    x = np.zeros(N)
    y = np.zeros(N)
    i = -1
    for row in f:
        if i >= 0:   # First row is label
            a, b = row.split(" ")
            x[i] = float(a)
            y[i] = float(b)
            #print(x[i])
            i = i + 1
        else:
            i = 0
        if i > imax:
            break
    ax.plot(x[0:min(i,imax)], y[0:min(i,imax)],label=label)
ax.legend()
fig
```
